```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial import ConvexHull
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import (
    KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering,
    MeanShift, OPTICS, Birch, estimate_bandwidth
)
from sklearn.mixture import GaussianMixture
from sklearn.metrics import (
    silhouette_score, davies_bouldin_score, calinski_harabasz_score,
    adjusted_rand_score, normalized_mutual_info_score
)
from scipy.cluster.hierarchy import dendrogram, linkage

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (8, 6)
np.random.seed(42)

pd.set_option('display.float_format', lambda x: f'{x:.3f}')
pd.set_option('display.max_columns', None)

# load the data
iris = load_iris()
X = iris.data
y_true = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

df = pd.DataFrame(X, columns=feature_names)
df['species'] = pd.Categorical.from_codes(y_true, target_names)
print(df.shape)
print(df.head())

sns.pairplot(df, hue='species', diag_kind='kde', palette='viridis')
plt.suptitle('Iris features by species')
plt.show()

# scale everything, then squish to 2D just so we can plot it
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
print('variance kept by 2 PCs:', round(pca.explained_variance_ratio_.sum() * 100, 1), '%')

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_true, cmap='viridis', s=60, edgecolor='k')
plt.xlabel('PC1'); plt.ylabel('PC2')
plt.title('True species (for reference only, not used by any algorithm)')
plt.legend(handles=scatter.legend_elements()[0], labels=list(target_names))
plt.show()


results = []


def evaluate(name, labels, n_outliers=0):
    mask = labels != -1  # DBSCAN/OPTICS mark noise as -1, exclude it from scoring
    n_clusters = len(set(labels[mask])) if mask.sum() > 0 else 0

    if n_clusters > 1:
        sil = silhouette_score(X_scaled[mask], labels[mask])
        db = davies_bouldin_score(X_scaled[mask], labels[mask])
        ch = calinski_harabasz_score(X_scaled[mask], labels[mask])
    else:
        sil = db = ch = np.nan

    ari = adjusted_rand_score(y_true, labels)
    nmi = normalized_mutual_info_score(y_true, labels)
    n_noise = int((labels == -1).sum())

    results.append({
        'Method': name, 'Clusters': n_clusters, 'Noise': n_noise, 'Outliers': int(n_outliers),
        'Silhouette': sil, 'DB Index': db, 'CH Score': ch, 'ARI': ari, 'NMI': nmi
    })
    print(f'{name}: {n_clusters} clusters, silhouette {sil:.3f}, ARI {ari:.3f}')


def cluster_geometry(labels):
    # centroid + radius for each cluster, and flags anything that's more
    # than 2 std devs away from its own cluster's centroid as an outlier
    outlier_flags = np.zeros(len(labels), dtype=bool)
    rows = []

    for lab in sorted(set(labels) - {-1}):
        idx = np.where(labels == lab)[0]
        pts = X_scaled[idx]
        centroid = pts.mean(axis=0)
        centroid_real = scaler.inverse_transform([centroid])[0]
        dist = np.linalg.norm(pts - centroid, axis=1)

        if dist.std() > 0:
            z = (dist - dist.mean()) / dist.std()
            outlier_flags[idx[z > 2]] = True

        rows.append({
            'Cluster': lab, 'Size': len(idx),
            'Avg radius': round(dist.mean(), 2), 'Max radius': round(dist.max(), 2)
        })
        print(f'  cluster {lab} (n={len(idx)}) centroid ->',
              ', '.join(f'{f.split()[0]}={v:.2f}' for f, v in zip(feature_names, centroid_real)))

    outlier_flags[labels == -1] = True  # DBSCAN/OPTICS noise counts as outliers too
    return pd.DataFrame(rows), outlier_flags


def plot_clusters(name, labels, outliers=None):
    plt.figure(figsize=(7, 5.5))
    for lab in sorted(set(labels)):
        mask = labels == lab
        if lab == -1:
            plt.scatter(X_pca[mask, 0], X_pca[mask, 1], s=55, color='lightgray', edgecolor='k', label='noise')
            continue

        pts = X_pca[mask]
        plt.scatter(pts[:, 0], pts[:, 1], s=55, edgecolor='k', label=f'cluster {lab}')

        if len(pts) >= 3:
            try:
                hull = ConvexHull(pts)
                boundary = np.vstack([pts[hull.vertices], pts[hull.vertices[0]]])
                plt.plot(boundary[:, 0], boundary[:, 1], linewidth=1.2)
                plt.fill(boundary[:, 0], boundary[:, 1], alpha=0.1)
            except Exception:
                pass  # too few / collinear points to draw a hull, just skip it

        plt.scatter(*pts.mean(axis=0), marker='x', s=120, color='black')

    if outliers is not None and outliers.any():
        plt.scatter(X_pca[outliers, 0], X_pca[outliers, 1], facecolors='none',
                    edgecolors='red', s=160, label='outlier')

    plt.title(name)
    plt.xlabel('PC1'); plt.ylabel('PC2')
    plt.legend()
    plt.show()


# K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_kmeans = kmeans.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_kmeans)
print(geom)
evaluate('K-Means', labels_kmeans, out.sum())
plot_clusters('K-Means', labels_kmeans, out)

# Hierarchical
linked = linkage(X_scaled, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(linked, truncate_mode='lastp', p=20)
plt.title('Dendrogram (Ward linkage)')
plt.show()

agglo = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels_agglo = agglo.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_agglo)
print(geom)
evaluate('Hierarchical', labels_agglo, out.sum())
plot_clusters('Hierarchical Clustering', labels_agglo, out)

# DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_dbscan)
print(geom)
evaluate('DBSCAN', labels_dbscan, out.sum())
plot_clusters('DBSCAN', labels_dbscan, out)

# Gaussian Mixture
gmm = GaussianMixture(n_components=3, random_state=42)
labels_gmm = gmm.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_gmm)
print(geom)
evaluate('GMM', labels_gmm, out.sum())
plot_clusters('Gaussian Mixture Model', labels_gmm, out)

# Spectral Clustering
spectral = SpectralClustering(n_clusters=3, random_state=42, affinity='nearest_neighbors', n_neighbors=10)
labels_spectral = spectral.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_spectral)
print(geom)
evaluate('Spectral', labels_spectral, out.sum())
plot_clusters('Spectral Clustering', labels_spectral, out)

# Mean Shift
bandwidth = estimate_bandwidth(X_scaled, quantile=0.2, random_state=42)
mean_shift = MeanShift(bandwidth=bandwidth)
labels_meanshift = mean_shift.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_meanshift)
print(geom)
evaluate('Mean Shift', labels_meanshift, out.sum())
plot_clusters('Mean Shift', labels_meanshift, out)

# OPTICS
optics = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.05)
labels_optics = optics.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_optics)
print(geom)
evaluate('OPTICS', labels_optics, out.sum())
plot_clusters('OPTICS', labels_optics, out)

# BIRCH
birch = Birch(n_clusters=3)
labels_birch = birch.fit_predict(X_scaled)
geom, out = cluster_geometry(labels_birch)
print(geom)
evaluate('BIRCH', labels_birch, out.sum())
plot_clusters('BIRCH', labels_birch, out)


results_df = pd.DataFrame(results).sort_values('Silhouette', ascending=False).reset_index(drop=True)
print(results_df)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
s = results_df.sort_values('Silhouette')
axes[0].barh(s['Method'], s['Silhouette'], color='steelblue')
axes[0].set_title('Silhouette score (higher = better)')

a = results_df.sort_values('ARI')
axes[1].barh(a['Method'], a['ARI'], color='coral')
axes[1].set_title('ARI vs true species (higher = better)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(results_df['Silhouette'], results_df['ARI'], s=90, edgecolor='k')
for _, row in results_df.iterrows():
    plt.annotate(row['Method'], (row['Silhouette'], row['ARI']), xytext=(5, 5), textcoords='offset points')
plt.xlabel('Silhouette (internal)'); plt.ylabel('ARI (external)')
plt.title('Do the internal and external scores agree?')
plt.show()

plt.figure(figsize=(8, 5))
o = results_df.sort_values('Outliers')
plt.barh(o['Method'], o['Outliers'], color='firebrick')
plt.title('Outliers flagged per method')
plt.tight_layout()
plt.show()

print('\nBest silhouette:', results_df.iloc[0]['Method'])
print('Best ARI:', results_df.sort_values('ARI', ascending=False).iloc[0]['Method'])
print('Fewest outliers:', results_df.sort_values('Outliers').iloc[0]['Method'])
```
